####Example of How to use Python To Calculate Certain Formula

```python
#Example of How to Calculate Area of Rectangle, Triangle, and parallelogram 
def Kalkulator_luas(p,l,L) : 
    luas_BujurSangkar = p*l
    luas_Segita = 0.5*p*l
    luas_JajarGenjang = p*l
    return luas_BujurSangkar, luas_Segita,luas_JajarGenjang
Kalkulator_Luas = Kalkulator_luas(5,4,0)
print ("Luas Bujur Sangkar", Kalkulator_Luas[0] ,"cm^2")
print ("Luas Segitiga", Kalkulator_Luas[1] ,"cm^2")
print ("Luas Jajar Genjang", Kalkulator_Luas[2] ,"cm^2")

def Kalkulator_Trapesium(a,b,t) :
    luas_Trapesium = 0.5*t*(a+b)
    return luas_Trapesium
kalkulatorTrapesium = Kalkulator_Trapesium (2,4,3)
print ("Luas Trapesium", kalkulatorTrapesium ,"cm^2")
```

####Example of How to Use Python to Process Data 

```python
#Import file data banking
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df_Cbanking = pd.read_csv(r"E:\MINE\beasiswa\2022\DATA BANGLORE\ON BOARDING\EXAM + TASK\3 - DATA PREPROCESSING\DATA TASK\banking.csv")
df_Cbanking
```

```python
#Checking data information
df_Cbanking.info()
```

```python
#checking missing data
df_Cbanking.isnull().sum()
```

```python
#Transform data, deleting unused column

df_Cbanking.drop(['emp_var_rate', 'cons_price_idx', 'cons_conf_idx', 'euribor3m', 'nr_employed', 'y'], axis = 1)
```

```python
#Filtering useful data
dfCbanking = df_Cbanking[['job', 'education', 'duration']]
dfCbanking
```

```python
#Checking categorical data in 'job' column
dfCbanking['job'].value_counts()
```

```python
#Checking categorical data in 'education' column
dfCbanking['education'].value_counts()
```

```python
#Labelling data 

list_status = []

for i, j in dfCbanking.iterrows():
    if ((j['job'] == 'management') and (j['education'] == 'university.degree') and (j['duration'] <=150)) :
        segment = 'rich'
    elif ((j['job'] == 'technician') and (j['education'] == 'university.degree') and (j['duration'] >= 150 and j['duration'] <= 200)):
        segment = 'middle'
    elif ((j['job'] == 'blue-collar') and (j['education'] == 'basic.4y' or j['education'] == 'basic.6y' or j['education'] == 'basic.9y') and (j['duration'] >200)) :
        segment = 'poor'
    else :
        segment = 'unknown'
    list_status.append(segment)
    
dfCbanking['Status'] = list_status
dfCbanking['Status'].value_counts()
```

```python
#Dropping unsed data
dfCbanking.drop(dfCbanking[dfCbanking['Status'] == 'unknown'].index)
```
