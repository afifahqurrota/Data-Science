### DATA PREPROCESSING

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
#Example of How to Create DataFrame with Dictionary Variable 

#Dictionary 1,2,3:
PialaEropa = {"Negara" : ["jerman", "spanyol", "uni soviet", "italia", "prancis"], 
              "piala": [3, 3, 1, 1, 2] }
PialaAsia = {"Negara" : ["jepang", "indonesia", "korea selatan", "irak", "saudi arabia"], 
             "piala" : [4, 0, 2, 1, 3]}
PialaDunia = {"Negara" : ["jerman", "spanyol", "korea selatan", "jepang", "prancis"],
             "piala" : [4, 1, 0, 0, 3]}

#Membuat DataFrame dari Dictionary 1,2,3:
dfPialaEropa= pd.DataFrame.from_dict(PialaEropa)
dfPialaAsia= pd.DataFrame.from_dict(PialaAsia)
dfPialaDunia= pd.DataFrame.from_dict(PialaDunia)
```

```python
dfPialaEropa
```

```python
dfPialaAsia
```

```python
dfPialaDunia
```

```python
#Example of How to Group Variables 
Pemenang = pd.concat([dfPialaEropa, dfPialaAsia, dfPialaDunia]).reset_index()
Pemenang
```

```python
#Example of How to Sum Up Row with the Same Variable
Total_Piala = Pemenang.groupby(['Negara']).sum()[['piala']]
Total_Piala.reset_index()
```

```python
#Example of How to Sort Values, False to sort Z-A, True to sort A-Z
juaraBertahan = Total_Piala.sort_values(by='piala', ascending = False).reset_index()
juaraBertahan
```

```python
#Example of How to Show Variable Index Value

JuaraBertahan = juaraBertahan.iloc[[0,3,5]].reset_index()
JuaraBertahan
```

```python
#Example of How to Filter Conditional Data
Pemenang[(Pemenang['piala'] >= 2)]
```

#### Example of Data Preprocessing Implementation

```python
#Example of How to Import File

dfCreditScore = pd.read_csv(r"E:\MINE\beasiswa\2022\DATA BANGLORE\ON BOARDING\EXAM + TASK\3 - DATA PREPROCESSING\DATA TASK\credit_score.csv", sep = ',')
dfCreditScore
```

```python
#Checking Data Type 
dfCreditScore.info()
```

```python
#Checking missing data
dfCreditScore.isnull().sum()
```

```python
#Checking statistical description of the data
dfCreditScore.describe()
```

```python
#Handle missing data 

dfCreditScore['umur']=dfCreditScore['umur'].fillna(dfCreditScore['umur'].median())
dfCreditScore['pendapatan']=dfCreditScore['pendapatan'].fillna(dfCreditScore['pendapatan'].median())
dfCreditScore['family_member']=dfCreditScore['family_member'].fillna(dfCreditScore['family_member'].median())
dfCreditScore.isnull().sum()
```

```python
#Changing Data Type of 'umur, family_member, pendapatan' 

dfCreditScore['umur'] = dfCreditScore['umur'].astype(int)
dfCreditScore['family_member'] = dfCreditScore['family_member'].astype(int)
dfCreditScore['pendapatan'] = dfCreditScore['pendapatan'].astype(int)
dfCreditScore
```

```python
#Filtering Conditional Data to Know Certain Customer Characteristics that Get Credit Access

Acc_Credit = dfCreditScore[~((dfCreditScore['umur'] > 40) & (dfCreditScore['family_member'] >= 3) & (dfCreditScore['pendapatan'] < 15000000))]
Acc_Credit
```

```python
#Filtering Conditional Data to Know Certain Customer Characteristics that Get Denied on Credit Access
Deny_PinjamanCS = dfCreditScore[((dfCreditScore['umur'] > 40) & (dfCreditScore['family_member'] >= 3) & (dfCreditScore['pendapatan'] < 15000000))]
Deny_PinjamanCS
```

```python
#Checking the correctness of the data, based on the amount of credit rejected and not rejected data
cekDataCS = pd.concat([Deny_PinjamanCS,Acc_Credit]).reset_index()
cekDataCS
```

```python
#Checking Statistical Description of Customer that Get Credit Access
Acc_Credit.describe()
```
